```python
#Load district shapefile
gh_districts_shapefile = r"C:\Users\FrancisAsanteNsiah\Desktop\gis_project\LERA_2425\Ghana  260 Districts_Boundary. 26.06.2019\260_Districts_of_Ghana.shp"
gh_districts = gpd.read_file(gh_districts_shapefile)
print(gh_districts)
```

```python
print(gh_districts.DISTRICT)
```

```python
#Plot the district shapefile
gh_districts.plot(column='DISTRICT', cmap='OrRd', legend=False, figsize=(12, 8))
plt.title("District Map of Ghana")
plt.show()
```

```python
ghana_shapefile = r"C:\Users\FrancisAsanteNsiah\Desktop\gis_project\LERA_2425\Ghana Shapefile_New\Ghana_Shapefile_New.shp"
ghana_regions = gpd.read_file(ghana_shapefile)

ghana_regions.plot(column='REGION', cmap='OrRd', legend=False, figsize=(12, 8))
plt.title("Regions Map of Ghana")
plt.show()
```

```python
#filter Western region
western_clipping = gh_districts[gh_districts['REGION'] == 'WESTERN']
#Clip Western region from the district map
western_region = gpd.clip(gh_districts, western_clipping)
western_region
```
